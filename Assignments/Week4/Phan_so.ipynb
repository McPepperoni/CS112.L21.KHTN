**Abstract:**<br>
Cho 2 phân số tối giản $a/b$ và $c/d$, mỗi bước tăng **a** và **b** thêm 1 đơn vị rồi tối giản phân số nhận được. Tìm số bước để đạt được phân số thứ 2<br>
**Decomposition:**<br>
Thực hiện đến khi đạt được phân số lớn hơn hoặc bằng phân số thứ 2 thì dừng<br>
**Patern Recognition:**<br>
Thực hiện vòng lập biến đổi phân số thứ nhất tới khi nhận được phân số lớn hơn hoặc bằng phân số thứ hai.<br>
Ta xét xem phân số mới này có bằng phân số thứ hai hay không, nếu có ta sẽ xuất ra số bước chạy vòng lập trên.<br>
Nếu điều kiện trên không thỏa, xuất ra giá trị -1 ( có nghĩa là không biến đổi được thành phân số thứ hai)

###**Algorithmic Design:**

```python
import math

a = int(input())
b = int(input())
c = int(input())
d = int(input())

cnt = 0

if not (c == d or (d*a-c*b)*(c-d)<0 or (d*a - c*b)%(c-d) != 0):
    while a != c or b != d:
        a = a + 1
        b = b + 1
        
        gdc = math.gcd(a, b)

        a = a // gdc
        b = b // gdc

        cnt = cnt + 1

        if cnt >= 1000000:
            cnt = 0
            break
    
print(cnt)
```
