```python
def timmax(a):
    n = len(a)
    if n == 1:
        return 0
    if n == 2:
      if a[0] > a[1]:
        return 0
      else:
        return 1

    mid = len(a) // 2
    left = a[0:mid]
    right = a[mid:len(a)]
    L = timmax(left)
    R = timmax(right) + mid
    if a[L] > a[R]:
      return L
    else:
      return R

a = input().split()
print(a)
print("Vị trí phần tử lớn nhất: "+str(timmax(a)))
```

### Recurrence Relation
$T(n) = T(n/2) + n$

$O(n) = nlog(n)$

###So sánh với thuật toán Brute Force
- Ưu điểm: Thời gian xử lí nhanh hơn
- Nhược điểm: Chiếm nhiều bộ nhớ hơn
